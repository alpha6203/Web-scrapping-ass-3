```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import requests
from PIL import Image
from io import BytesIO

# Function to create a directory if it doesn't exist
def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

# Function to save image from URL
def save_image(url, directory, filename):
    try:
        response = requests.get(url)
        img = Image.open(BytesIO(response.content))
        img.save(os.path.join(directory, filename))
    except Exception as e:
        print(f"Error occurred while saving image: {e}")

# Function to scrape images
def scrape_images(keyword, num_images):
    # Launching Chrome browser using Selenium
    driver = webdriver.Chrome(executable_path='/path/to/chromedriver')  # Replace with the path to your chromedriver
    
    # Navigating to Google Images
    driver.get('https://www.google.com/imghp?hl=en')
    
    # Locating the search bar and search button
    search_bar = driver.find_element_by_name('q')
    search_button = driver.find_element_by_xpath('//button[@type="submit"]')
    
    # Entering keyword and searching
    search_bar.send_keys(keyword)
    search_button.click()
    
    # Waiting for the results to load
    time.sleep(3)
    
    # Scrolling to load more images
    for _ in range(3):
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)
    
    # Scraping image URLs
    image_urls = []
    for i in range(num_images):
        try:
            # Locating image elements
            image_elements = driver.find_elements_by_xpath('//img[@class="rg_i Q4LuWd"]')
            image_url = image_elements[i].get_attribute('src')
            if image_url and 'http' in image_url:
                image_urls.append(image_url)
        except Exception as e:
            print(f"Error occurred while scraping image URL: {e}")
    
    # Closing the browser
    driver.quit()
    
    # Creating a directory to save images
    directory = f'{keyword}_images'
    create_directory(directory)
    
    # Saving images
    for i, url in enumerate(image_urls):
        filename = f'{keyword}_{i+1}.jpg'
        save_image(url, directory, filename)
    
    print(f"{num_images} images scraped for '{keyword}'.")

# Keywords and number of images to scrape
keywords = ['fruits', 'cars', 'Machine Learning', 'Guitar', 'Cakes']
num_images_per_keyword = 10

# Scraping images for each keyword
for keyword in keywords:
    scrape_images(keyword, num_images_per_keyword)
```
