```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def scrape_forbes_billionaires():
    url = 'https://www.forbes.com/billionaires/'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    billionaires = []
    
    for row in soup.find_all('div', class_='personName'):
        billionaire_details = {}
        
        try:
            billionaire_details['Rank'] = row.find_previous('div', class_='rank').text.strip()
        except:
            billionaire_details['Rank'] = '-'
        
        try:
            billionaire_details['Name'] = row.find('div', class_='name').text.strip()
        except:
            billionaire_details['Name'] = '-'
        
        try:
            billionaire_details['Net Worth'] = row.find_next('div', class_='netWorth').text.strip()
        except:
            billionaire_details['Net Worth'] = '-'
        
        try:
            billionaire_details['Age'] = row.find_next('div', class_='age').text.strip()
        except:
            billionaire_details['Age'] = '-'
        
        try:
            billionaire_details['Citizenship'] = row.find_next('div', class_='countryOfCitizenship').text.strip()
        except:
            billionaire_details['Citizenship'] = '-'
        
        try:
            billionaire_details['Source'] = row.find_next('div', class_='source-column').text.strip()
        except:
            billionaire_details['Source'] = '-'
        
        try:
            billionaire_details['Industry'] = row.find_next('div', class_='category').text.strip()
        except:
            billionaire_details['Industry'] = '-'
        
        billionaires.append(billionaire_details)
    
    return billionaires

# Scrape data
billionaires = scrape_forbes_billionaires()

# Create DataFrame
df = pd.DataFrame(billionaires)

# Save to CSV
df.to_csv('forbes_billionaires.csv', index=False)
```
